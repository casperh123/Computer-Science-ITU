```python
# Write your solution here
print(A @ C)
print(C @ A)

# The two result are diffent, since matrix multiplication is not commutitative.
# With A  @ C, we take the dot product of the rows of A with the columns of C.
# With C @ C we take the dot product of the rows of C with the columns of A

print(D @ E)
print(E @ D)

#E @ D fail because the inner dimensions of the matrices are different 
# E is 3x3 and D is 2 x 3. Thus when E is on the left side of the matrix multiplication, the colunmns of E are projected onto the rows of D, thus transforming D from R^3 -> R^2. The two rows of E act as 2 linear
# transformations on the columns of E
```
